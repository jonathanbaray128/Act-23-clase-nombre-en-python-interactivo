```python
class Contador:
    def __init__(self, nombre_contador, valor_inicial=0, incremento_paso=1):
        self.nombre = nombre_contador
        self.valor = valor_inicial
        self.paso = incremento_paso

    def incrementar(self):
        self.valor += self.paso
        print(f"➕ Incrementado. Valor actual: {self.valor}")

    def decrementar(self):
        self.valor -= self.paso
        print(f"➖ Decrementado. Valor actual: {self.valor}")

    def resetear(self):
        self.valor = 0
        print(f"🔄 Contador reiniciado a 0.")

    def mostrar_estado(self):
        print(f"\n--- Panel del Contador ---")
        print(f"Nombre: {self.nombre}")
        print(f"Valor Actual: {self.valor}")
        print(f"Ritmo de cambio: {self.paso}")
        print(f"--------------------------")

# --- Lógica de Interacción ---

# Creamos una instancia de ejemplo
mi_contador = Contador("Contador de Pasos", 0, 1)

print(f"¡Sistema de Conteo '{mi_contador.nombre}' activado!")

while True:
    print("\n¿Qué deseas hacer?")
    print("1. Ver valor actual")
    print("2. Incrementar (+)")
    print("3. Decrementar (-)")
    print("4. Resetear")
    print("5. Salir")

    opcion = input("Selecciona una opción (1-5): ")

    if opcion == "1":
        mi_contador.mostrar_estado()

    elif opcion == "2":
        mi_contador.incrementar()

    elif opcion == "3":
        mi_contador.decrementar()

    elif opcion == "4":
        mi_contador.resetear()

    elif opcion == "5":
        print("Saliendo del contador... ¡Hasta pronto!")
        break

    else:
        print("❌ Opción no válida, intenta de nuevo.")
```
